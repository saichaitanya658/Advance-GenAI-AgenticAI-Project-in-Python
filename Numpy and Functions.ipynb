# array manupulation functions

```python
a = np.array([1,2,3,4,5])
a
```

```python
reshaped = np.arange(0,10).reshape(2,5)
print("reshaped array is:\n",reshaped)
```

```python
a1 = np.array([[1,2],[3,4],[5,6],[7,8]])
flattened = np.ravel(a1)
print("flattened array is:\n",flattened)
```

```python
a2 = np.array([[1,5],[4,8]])
transposed = np.transpose(a2)
print("transposed array is :\n",transposed)
```

```python
a3 = np.array([1,3])
a4 = np.array([2,4])
stacked = np.vstack([a3,a4])
print("vertically stacked array :\n",stacked)
```

```python
a3 = np.array([1,3])
a4 = np.array([2,4])
stacked = np.hstack([a3,a4])
print("horizontal stacked array :\n",stacked)
```
